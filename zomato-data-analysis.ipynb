```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path='https://raw.githubusercontent.com/TheiScale/YouTube-Video-Notes/refs/heads/main/Zomato_Python_Project/Zomato%20data%20.csv'
data=pd.read_csv(path)
data.head()
```

```python
sns.countplot(x='listed_in(type)',data=data,palette="tab10")
plt.show()
```

```python
a=data.groupby('listed_in(type)')['votes'].sum()
a=a.reset_index()
a.columns=['cafe_type','votes']
print(a)
sns.barplot(x='cafe_type',y='votes',palette='tab10',data=a)
# plt.show()
```

```python
b=data['rate'].value_counts()
b=b.reset_index()
b.columns=['rating','total']
```

```python
sns.barplot(x='rating',y='total',data=b,palette='tab10')
plt.show()
```

```python
c=data['approx_cost(for two people)'].value_counts()
c=c.reset_index()
c.columns=['costs','number']
c
```

```python
sns.barplot(x='costs',y='number',data=c,palette='tab10')
plt.show()
```
